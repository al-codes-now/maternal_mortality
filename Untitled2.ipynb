```python
import pandas as pd
import statsmodels.api as sm
from statsmodels.formula.api import ols
import matplotlib.pyplot as plt
plt.style.use('seaborn')
```

```python
df = pd.read_csv('/Users/alexandrabruno/Documents/flatiron/data/final_df.csv')
```

```python
mm = pd.read_csv('/Users/alexandrabruno/Documents/flatiron/data/test.csv')
```

```python
# Get indexes where name column has value PWGTP
indexNames = mm[mm['cause'] == 'All causes'].index
 
# Delete these row indexes from dataFrame
mm.drop(indexNames, inplace=True)
```

```python
mm.drop(['sex', 'upper','lower'], axis = 1,inplace = True) 
```

```python
mm = mm.rename(columns = {'location':'STATE','age':'AGE','year':'YEAR','val':'DEATH_RTE'})
```

```python
df = pd.merge(mm, df,  how='left', left_on=['STATE','YEAR','AGE'], right_on = ['STATE','YEAR','AGE'])
```

```python
df.dropna(inplace = True)
```
