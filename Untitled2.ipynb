```python
# Get indexes where name column has value PWGTP
indexNames = mm[mm['cause'] == 'All causes'].index
 
# Delete these row indexes from dataFrame
mm.drop(indexNames, inplace=True)
```

```python
del mm['upper']
del mm['measure']
del mm['sex']
del mm['lower']
```

```python
mm = mm.rename(columns = {'location':'STATE','age':'AGE','year':'YEAR'})
```

```python
mm = mm.rename(columns = {'val':'DEATH_RTE'})
```

```python
df = pd.merge(mm, df,  how='left', left_on=['STATE','YEAR','AGE'], right_on = ['STATE','YEAR','AGE'])
```

```python
df.dropna(inplace = True)
```
