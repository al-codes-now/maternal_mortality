```python
# for year in range of years:
# replace year with next year

def get_data(url,start_year, stop_year):
    data_frames = []
    for year in range(start_year,stop_year):
        url = url.replace('2004',str(year))
        response = requests.get(url)
        data = response.text
        data = json.loads(data)
        df = pd.DataFrame.from_dict(data)
        df['YEAR'] = df.apply(lambda x: year, axis=1)
        data_frames.append(df)
    
    return data_frames
```

```python
dfs = get_data(url,2004,2020)
```
